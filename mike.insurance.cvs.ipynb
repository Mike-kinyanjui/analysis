```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data=pd.read_csv(r"C:\Users\mike\Downloads\insurance.csv")
```

```python
print(data)
```

```python
#checking for missing values

data.isnull().sum()
```

```python
#checking the datatype of each attribute
data.info()
```

```python
#checking for wrong entry symbols
for col in data.columns:
    print('{} : {}'.format(col,data[col].unique()))
```

```python
#checking correlation between different variables
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(),cbar=True,annot=True,cmap='Greens')
```

```python
#relationship between charges and age
plt.figure(figsize=(10,10))
plt.scatter(x='age',y='charges',data=data)
plt.xlabel('age')
plt.ylabel('charges')
```

```python
 #univariate analysis of age
 sns.histplot(data.age,bins=10)
```

```python
#relationship btn age and sex
sns.boxplot(x='age',y='sex',data=data)
```
